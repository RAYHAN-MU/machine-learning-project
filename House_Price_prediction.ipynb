Linear Regression With Single Variable

```python
import pandas as pd
import numpy as np

np.random.seed(0)
areas = np.random.randint(5000, 20000, size=50)
prices = 12000 + (areas - 10000) * 1.5

# Create a DataFrame
df = pd.DataFrame({
    'area': areas,
    'price': prices
})

# Save to CSV
df.to_csv('housingprice.csv', index=False)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('housingprice.csv')
df.tail()
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt
plt.xlabel('area(sq(feet))')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.price,color='red')
```

```python
model=LinearRegression()
model.fit(df[['area']],df.price)
```

```python
model.predict([[3200]])
```
